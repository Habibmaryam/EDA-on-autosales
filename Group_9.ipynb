# **sales of the top 5 customers across different countries; Most Loyal Customer**

```python
# Calculating and displaying the top 5 customers in each country based on total sales in the DataFrame "autosales"
# Grouping the DataFrame by country and customer name, summing the sales for each combination.
# Getting the top 5 customers in each country based on total sales.
top_customers_by_country = autosales.groupby(['COUNTRY', 'CUSTOMERNAME'])['SALES'].sum().reset_index()

# Using groupby and nlargest to get the top 5 customers in each country based on total sales
top_5_customers_by_country = top_customers_by_country.groupby('COUNTRY').apply(lambda x: x.nlargest(5, 'SALES')).reset_index(drop=True)

# Sorting the overall result to obtain the top 5 customers across all countries
top_5 = top_5_customers_by_country.sort_values(by="SALES", ascending=False).head(5)

# Printing the top 5 customers by country
print(top_5)
```

```python

# Creating a bar plot to visualize the sales of the top 5 customers across different countries in the DataFrame "top_5"
# Using Seaborn's sns.barplot with Sales on the x-axis, Customer Name on the y-axis, and Country as hue.
sns.barplot(data=top_5, x='SALES', y='CUSTOMERNAME', hue='COUNTRY', palette='plasma')

# Adding a legend to the plot
plt.legend(loc='lower right')

# Adding a title to the plot
plt.title("Most Loyal Customers", fontdict={"fontsize": 20})

# Show the plot
plt.show()
```

**Euroshopping brought in the most sales of about 700,000 which indicated that they are the most loyal customers while La Rochelle Gifts is the 5th most loyal customer with about 190,000 in sales with their office in France.**

## **Order Count Analysis By Day**

```python
#Extracting day from ORDERDATE column
autosales["DAY"] = autosales["ORDERDATE"].dt.day_name()
#coducting value counts analysis on the extracted variable
order_count_by_day = autosales["DAY"].value_counts()
#Defining the order of the date
days= ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]

#plotting order count by day barchart
plt.figure(figsize=(8,4))
sns.barplot(x=order_count_by_day.index , y = order_count_by_day.values, order= days, hue = order_count_by_day.index, legend = False)
#adding details to the plot
plt.title("Order Count By Day")
plt.xlabel("Days")
plt.ylabel("Number of Orders")
plt.show()
```

**In comparing the amount of orders received in each day's of the week. Orders placed on sunday's had the highest amount, with orders placed on thursday's having the least amount..**


# **RFM Analysis**

```python
#calculating recency, frequency and monetary values for each customer
rfm_data = autosales.groupby("CUSTOMERNAME").agg({
    "ORDERDATE": "max",
    "ORDERNUMBER":'nunique',
    "SALES" : "sum"
})

#calculating recency as the number of days since the order was placed
rfm_data["Recency"] = (pd.to_datetime("now")- rfm_data["ORDERDATE"]).dt.days

#sorting the data by the highest number of sales.
rfm_data_sorted = rfm_data.sort_values(by= "SALES", ascending = False)

rfm_data_sorted.head()
```

**The RFM analysis shows top 5 customers with the highest amount in sales, the number of orders they have placed, the last date they placed an order,and the number of days it has been since that order was placed**
